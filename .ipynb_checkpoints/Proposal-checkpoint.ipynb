### **2. Research Question**
**My research question is**: Do weather conditions influence investor behavior in financial markets?

The project will the examine whether daily weather conditions, such as temperature and cloud coverage, are associated with changes in stock market returns. 
### **3. Variables to be collected** 

**Variable 1**: Stock market returns 
Source: Alpha Vantage API

This variable will include daily stock prices for a selected market index (such as S&P 500).

Daily returns will be calculated as:
Return = (Price_t − Price_(t−1)) / Price_(t−1)

Where:
- Price_t is the closing price on day t
- Price_(t−1) is the closing price on the previous day


**Variable 2**: Weather Conditions Rate
Source: Tomorrow.io API

This variable includes daily weather indicators:
- Temperature
- Cloud cover
- Humidity
- Wind speed
These variables will be used to construct a combined weather measure.


**Variable 3**: Binary Good Weather Variable (calculated variable) 

A day will be classified as “good weather” if it satisfies the following conditions:
- Temperature is within a comfortable range (e.g., 15–25°C)
- Cloud cover is low (e.g., below 50%)
- Wind speed is low to moderate

If all conditions are met:
GoodWeather = 1
Otherwise:
GoodWeather = 0

### **4. Methodology** 
The analysis will use a regression model where:

**Dependent variable**: Stock Market Returns

**Independent variable**: Good Weather Index

The model can be written as:

**Stock Returns = β0 + β1 * Good Weather Index + ε**

### **6. Background and Motivation** 
Understanding the factors that influence financial markets is one of the important tasks in economics. It is commonly believed that investors make decisions only rationally. However, I believe that human decisions can also be influenced by external factors such as weather.

Weather conditions can affect people’s mood and well-being, which may lead to changes in their decisions. Since investors’ decisions influence financial markets, any changes in their behavior may be reflected in market dynamics.
Studying the relationship between weather conditions and investor behavior can help us understand which factors, besides economic and political indicators, may influence financial markets.
